Assignment 03 – CSO Dataset Download

This program retrieves the Exchequer Account (Historical Series) dataset from the Central Statistics Office (CSO) API and saves it locally as a JSON file.

```python
# importing Python Libraries

import requests
import json

url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"

response = requests.get(url)

response.raise_for_status()

# get JSON data
data = response.json()

# save to file
with open("cso.json", "w", encoding="utf-8") as f:
    import json
    json.dump(data, f, ensure_ascii=False, indent=2)

print("Dataset saved to cso.json")
```
