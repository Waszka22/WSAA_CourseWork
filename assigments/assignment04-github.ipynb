### Assignment 04

Write a program in python that will read a file from a repository, 

The program should then replace all the instances of the text "Andrew" with your name. 

The program should then commit those changes and push the file back to the repository.


```python
import os
import subprocess
```

```python
# repository and file configuration
repo_url = "https://github.com/Waszka22/WSAA_CourseWork"
local_repo_folder = "WSAA_CourseWork"
file_to_edit = "text.txt"
old_text = "Andrew"
new_text = "Agnieszka"
```

```python
# function to run git commands
def run_git_command(command):
    result = subprocess.run(command, capture_output=True, text=True)
    if result.returncode != 0:
        print("Git Error", result.stderr.strip())
    return result

def main():
    # clone repository and pull latest changes
    if not os.path.exists(local_repo_folder):
        print("Cloning Repository")
        run_git_command(["git", "clone", repo_url, local_repo_folder])
    else:
        print("Pulling latest changes")
        run_git_command(["git", "-C", local_repo_folder, "pull"])

    file_path = os.path.join(local_repo_folder, file_to_edit)

    # create the file
    if not os.path.exists(file_path):
        with open(file_path, "w", encoding="utf-8") as f:
            f.write("")

    # read file content and replace old text with new text
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()
    updated_content = content.replace(old_text, new_text)
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(updated_content)

    # add commit and push changes
    run_git_command(["git", "-C", local_repo_folder, "add", file_to_edit])
    run_git_command([
        "git",
        "-C",
        local_repo_folder,
        "commit",
        "-m",
        f"Replaced {old_text} with {new_text}"
    ])
    run_git_command(["git", "-C", local_repo_folder, "push"])

    print("Done!")

if __name__ == "__main__":
    main()
```
