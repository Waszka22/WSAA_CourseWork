Deck of Cards API    https://deckofcardsapi.com/

This is an API that simulates dealing a deck of cards

Write a program that "deals" (prints out) 5 cards

```python
# # importing Python Libraries
import requests

# shuffle a new deck
shuffle_url = "https://deckofcardsapi.com/api/deck/new/shuffle/?deck_count=1"
shuffle_response = requests.get(shuffle_url)
shuffle_data = shuffle_response.json()

deck_id = shuffle_data["deck_id"]
```

```python
# draw 5 cards
draw_url = f"https://deckofcardsapi.com/api/deck/{deck_id}/draw/?count=5"
draw_response = requests.get(draw_url)
draw_data = draw_response.json()

cards = draw_data["cards"]

print("Your 5 cards are:\n")

values = []
suits = []

for card in cards:
    value = card["value"]
    suit = card["suit"]
    print(f"{value} of {suit}")
    values.append(value)
    suits.append(suit)

# convert face cards to numbers for straight checking
value_map = {
    "ACE": 14,
    "KING": 13,
    "QUEEN": 12,
    "JACK": 11
}

numeric_values = []
for v in values:
    if v.isdigit():
        numeric_values.append(int(v))
    else:
        numeric_values.append(value_map[v])



# check for pair or three of a Kind
value_counts = {}
for v in values:
    value_counts[v] = value_counts.get(v, 0) + 1

has_pair = False
has_three = False

for count in value_counts.values():
    if count == 2:
        has_pair = True
    if count == 3:
        has_three = True

# check for flush
has_flush = len(set(suits)) == 1

# check for straight
has_straight = False
if numeric_values == list(range(numeric_values[0], numeric_values[0] + 5)):
    has_straight = True

# print result and congratulate the user
print("\nResult:")
if has_three:
    print("Congratulations! You  Three of a Kind!")
elif has_pair:
    print("Congratulations! You have a Pair!")
if has_straight:
    print("Congratulations! You have a Straight!")
if has_flush:
    print("Congratulations! You have a Flush!")
    

if not (has_pair or has_three or has_straight or has_flush):
    print("Try again!")
```
